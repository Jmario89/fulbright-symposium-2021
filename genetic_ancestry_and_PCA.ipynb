```r
options(scipen=100, digits=3)
library('dplyr')
library('ggplot2')
```

```r
eigenvec <- read.table('data/allCombinedLDPruned.eigenvec', header = TRUE, skip=0, sep = ' ')
rownames(eigenvec) <- eigenvec[,2]
eigenvec <- eigenvec[,3:ncol(eigenvec)]
colnames(eigenvec) <- paste('PC', c(1:20), sep = '')
```

```r
eigenvec = eigenvec[c('PC1', 'PC2')]
eigenvec
```

```r
PED <- read.table('data/allCombined.ped', header = TRUE, skip = 0, sep = '\t')
PED <- PED[which(PED$Individual.ID %in% rownames(eigenvec)), ]
PED <- PED[match(rownames(eigenvec), PED$Individual.ID),]
all(PED$Individual.ID == rownames(eigenvec)) == TRUE
```
